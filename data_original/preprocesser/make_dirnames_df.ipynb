```python
from glob import glob
import os
import pandas as pd
import numpy as np
import re
```

```python
path = "../dogbone_dumbbell_A12/dogbone_dumbbell_A12/*"
new_pathes = glob(path)
new_pathes = sorted(new_pathes, key=lambda s: float(re.sub(r'[^\d.]', "", s.split("/")[-1])))
times = [float(path.split("/")[-1]) for path in new_pathes]
indexes = range(0, len(new_pathes))
df = pd.DataFrame(indexes, columns=["fileIndex"])
df["dirpath"] = new_pathes
df["isInitial"] = False
df["hasCoordinate"] = False
df["time"] = times
diffs = np.zeros_like(np.array(times))
diffs[1:] = np.array(times)[1:] - np.array(times)[:-1]
diffs[0] = 0.01
diffs = np.round(diffs, decimals=2)
df["dt"] = diffs

df.to_csv("dirnames_df.csv")
df
```
